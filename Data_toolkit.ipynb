THEORY QUESTIONS

Q1. What is NumPy, and why is it widely used in Python?

Answer:
NumPy is a Python library used for numerical and scientific computing. It provides support for large, multi-dimensional arrays and mathematical functions that operate efficiently on these arrays. It is widely used because it is fast, memory-efficient, and supports vectorized operations.

Q2. How does broadcasting work in NumPy?

Answer:
Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically expanding the smaller array to match the larger one without copying data.

Q3. What is a Pandas DataFrame?

Answer:
A Pandas DataFrame is a two-dimensional labeled data structure similar to a table, with rows and columns. It can store different data types and supports data manipulation and analysis.

Q4. Explain the use of the groupby() method in Pandas.

Answer:
The groupby() method is used to group data based on one or more columns and perform aggregate operations like sum, mean, or count on each group.

Q5. Why is Seaborn preferred for statistical visualizations?

Answer:
Seaborn provides attractive default styles and built-in statistical plots. It simplifies complex visualizations and works seamlessly with Pandas DataFrames.

Q6. What are the differences between NumPy arrays and Python lists?

Answer:
NumPy arrays are faster, consume less memory, and support vectorized operations, while Python lists are slower and store elements of different data types.

Q7. What is a heatmap, and when should it be used?

Answer:
A heatmap is a graphical representation of data where values are shown using colors. It is commonly used to visualize correlation matrices or intensity of values.

Q8. What does the term “vectorized operation” mean in NumPy?

Answer:
Vectorized operations allow calculations on entire arrays without using loops, making the code faster and more efficient.

Q9. How does Matplotlib differ from Plotly?

Answer:
Matplotlib is mainly used for static plots, while Plotly provides interactive and dynamic visualizations suitable for web applications.

Q10. What is the significance of hierarchical indexing in Pandas?

Answer:
Hierarchical indexing allows multiple levels of indexing in a DataFrame, making it easier to work with complex and multi-dimensional data.

Q11. What is the role of Seaborn’s pairplot() function?

Answer:
pairplot() creates pairwise plots between all numerical columns to analyze relationships and distributions in a dataset.

Q12. What is the purpose of the describe() function in Pandas?

Answer:
The describe() function provides summary statistics such as mean, median, count, min, max, and standard deviation.

Q13. Why is handling missing data important in Pandas?

Answer:
Handling missing data ensures accurate analysis and prevents incorrect results in data processing and modeling.

Q14. What are the benefits of using Plotly for data visualization?

Answer:
Plotly offers interactive charts, zooming, tooltips, and supports web-based dashboards.

Q15. How does NumPy handle multidimensional arrays?

Answer:
NumPy supports multidimensional arrays using the ndarray object, allowing efficient storage and operations on matrices and tensors.

Q16. What is the role of Bokeh in data visualization?

Answer:
Bokeh is used to create interactive visualizations for modern web browsers using Python.

Q17. Explain the difference between apply() and map() in Pandas.

Answer:
map() works on a single Series, while apply() can be used on Series or DataFrames to apply custom functions.

Q18. What are some advanced features of NumPy?

Answer:
Broadcasting, linear algebra operations, FFT, random number generation, and memory optimization.

Q19. How does Pandas simplify time series analysis?

Answer:
Pandas provides datetime indexing, resampling, shifting, and rolling window functions for time series data.

Q20. What is the role of a pivot table in Pandas?

Answer:
A pivot table summarizes data by aggregating values based on categories.

Q21. Why is NumPy’s array slicing faster than Python’s list slicing?

Answer:
NumPy slicing creates views instead of copies, making it faster and more memory-efficient.

Q22. What are some common use cases for Seaborn?

Answer:
Statistical plots, distribution analysis, correlation visualization, and exploratory data analysis.
